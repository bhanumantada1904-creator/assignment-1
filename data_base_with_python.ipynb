```python
import sqlite3

conn = sqlite3.connect(':memory:') # Connect to an in-memory SQLite database
cursor = conn.cursor()

create_table_sql = """
CREATE TABLE Books (
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,
    price REAL NOT NULL,
    stock_quantity INTEGER DEFAULT 0
);
"""

try:
    cursor.execute(create_table_sql)
    conn.commit()
    print("Table 'Books' created successfully.")
except sqlite3.Error as e:
    print(f"An error occurred: {e}")
finally:
    conn.close()
```

```python
import sqlite3

conn = sqlite3.connect(':memory:') # Connect to an in-memory SQLite database
cursor = conn.cursor()

create_table_customers_sql = """
CREATE TABLE Customers (
    customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    city TEXT,
    join_date TEXT
);
"""

try:
    cursor.execute(create_table_customers_sql)
    conn.commit()
    print("Table 'Customers' created successfully.")
except sqlite3.Error as e:
    print(f"An error occurred: {e}")
finally:
    conn.close()
```

```python
import sqlite3

# Re-establish connection for this cell, assuming previous connections were closed
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

create_table_orders_sql = """
CREATE TABLE Orders (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER,
    book_id INTEGER,
    order_date TEXT,
    quantity INTEGER,
    FOREIGN KEY (customer_id) REFERENCES Customers(customer_id),
    FOREIGN KEY (book_id) REFERENCES Books(book_id)
);
"""

try:
    cursor.execute(create_table_orders_sql)
    conn.commit()
    print("Table 'Orders' created successfully.")
except sqlite3.Error as e:
    print(f"An error occurred: {e}")
finally:
    conn.close()
```
