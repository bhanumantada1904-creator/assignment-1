```python
import pandas as pf
#employee data
data = {"Employee_ID":[201,202,203,204,205,206,207,208,209,210],
        "Name": ["Sarah","Michael","jennifer","David","Lisa","James","emma","robert","sophia","william"],
        "department": ["engineering","sales","Marketing","HR","engineering","sales","Marketing","HR","engineering","sales"],
        "salary":[85000,62000,71000,92000,58000,68000,78000,75000,65000,60000],
        "score":[4,3,5,5,4,4,3,4,5,3],
        "service":[5,3,7,8,4,6,2,5,4,3]
       }
df=pf.DataFrame(data)
print(df)
```

```python
df=pf.DataFrame(data)
selected_columns = df[['Name', 'department', 'salary']]
print(selected_columns)
```

```python
print(df.iloc[5])
```

```python
#set Employee_ID as index
df.set_index("Employee_ID",inplace=True)
#Display employee with Employee_ID 208
print(df.loc[208])
```

```python
engineering_employees=df[df["department"]=="engineering"]
print(engineering_employees)
```

```python
#Filter employees with salary greater than 70000
high_salary_employees=df[df["salary"]>70000]
print(high_salary_employees)
```

```python
#Filter employees from sales department with performance score >=4
filtered_employees=df[(df["department"]=="sales")&(df["score"]>=4)]
print(filtered_employees)
```

```python
#sort by salary in descending order
sorted_df=df.sort_values(by="salary",ascending=False)
print(sorted_df)
```
